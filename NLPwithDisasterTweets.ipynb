```python
train = pd.read_csv('/kaggle/input/nlp-getting-started/train.csv')
test = pd.read_csv('/kaggle/input/nlp-getting-started/test.csv')
```

```python
train = train.drop('id',axis=1)
train = train.drop('keyword',axis=1)
train = train.drop('location',axis=1)

```

```python
train = train.values
X_train = train[:,0]
Y_train = train[:,1]
```

```python

# 토큰화 함수를 이용해 전처리 하는 과정입니다.


token = Tokenizer()             # 토큰화 함수 지정
token.fit_on_texts(X_train)       # 토큰화 함수에 문장 적용
 
```

```python
x = token.texts_to_sequences(X_train)
```

```python
padded_x = pad_sequences(x, 100)
```

```python
word_size = len(token.word_index) +1
```

```python
model = Sequential()
model.add(Embedding(word_size, 8, input_length=100))  #벡터로 만들어준다. 숫자값으로 바꿔준다.
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
Y_train = Y_train.astype('float64')
```

```python
model.fit(padded_x, Y_train, epochs=20)
```

```python
test = test.drop('id',axis=1)
test = test.drop('keyword',axis=1)
test = test.drop('location',axis=1)

test = test.values
X_test = test[:,0]
```

```python
token1 = Tokenizer()  
token1.fit_on_texts(X_test)
y = token.texts_to_sequences(X_test)
padded_x_test = pad_sequences(y, 100)
```

```python
y_pre=model.predict(padded_x_test)
```
