## classfication 과 regression을 평가하는 모델들


### classfication
- metrics.accuracy_score(y_true, y_pred,*[,...]) 정확도 측정 모델
- metrics.average_precision_score(y_true,y_pred) 평균 정밀도 모델
- metrics.f1_score 데이터가 불균형할 때 쓰는 모델

### regression
- metrics.mean_absolute_error(y_true,y_pred) 리그레이션 절대값의 평균
- metrics.mean_squared_error(y_tre,y_pred) 평균의 제곱오차
- metrics.r2_score(y_true, y_pred) R^2 모델

```python
from keras.datasets import imdb

(x_train, y_train), (x_test, y_test) = imdb.load_data(
    num_words=10000)
```

```python
#numpy 는 list를 배열 인덱스로 넣어도 그 위치에만 다 들어감
def vectorize_sequences(sequences, dimension=10000):
    # 크기가 (len(sequences), dimension)이고 모든 원소가 0인 행렬을 만듭니다.
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = results[i,sequence]+1. # results[i]에서 특정 인덱스의 위치를 1로 만듭니다.
    return results
```

```python
x_train=vectorize_sequences(x_train)
x_test = vectorize_sequences(x_test)
```

```python
x_test = x_test.reshape(-1,10000,1)
x_train = x_train.reshape(-1,10000,1)
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,
                    epochs=20,
                    batch_size=1024, validation_split=0.2,verbose=2,callbacks=[early_stopping])
```
