# CNN 알고리즘

- 기존의 신경망은 완전연결로 이루어졌다.
- 모든 입력 데이터가 1차원이었는데 이런 경우 데이터의 형상을 무시한다는 단점이 있다.
- 이미지는 가로 28*세로 28 =784개의 데이터로 이루어 질수 있는데 이것을 한 차원의 784개의 데이터로 받는다는 것이다.
- 그렇다면 근접한 데이터는 비슷한 값을 가진다는 위치와 형상의 특성을 살릴 수 없다.


- 하지만 합성곱 계층은 3차원으로 입력데이터를 받으며 형상을 유지한다. 따라서 이미지처럼 형상을 가진 데이터를 제대로 이해하고
- CNN에서는 합성곱 계층의 입출력 데이터를 특징 맵이라고 한다.

- 신경망에서는 필터(정방행렬)의 매개변수가 가중치에 해당한다.
- 편향도 존재하는데 편향은 항상 하나만 존재한다. 그 하나의 값을 필터를 적용한 모든 원소에 더하는 것이다.


- 합성곱을 하게되면 출력 데이터의 크기가 작아지는데 합성곱을 연속으로 해야하는 신경망에서는 1*1 행렬이 되버리는 문제가 발생한다.
- 이를 해결하기 위한 기술이 패딩이라 한다. 입력 데이터에다가 특정 크기만큼 크기를 증가해 출력데이터의 크기를 조절하는 것이다.


- 스트라이드는 입력데이터와 필터의 합성곱을 1칸씩할지, 2칸씩할지, 정하는 것이다.


- 3차원 합성곱에서의 주의할 점은 입력 데이터의 채널 수 와 필터의 채널 수가 같아야 한다는 점이다.
- 하나의 필터당 하나의 차원에 합성곱을 진행하기 때문이다. 단, 필터의 크기는 원하는 값으로 설정이 가능하다.


- 3차원 합성곱에서 출력 데이터가 여러개 만드는 방법은 3차원 가중치 필터를 여러 개 만드는 것이다.
- 즉, 필터를 4차원까지 늘리는 것이다.
- 따라서 필터는 4차원이다 (출력 채널 수, 입력 채널 수, 높이, 너비) 출력 채널 수란 몇 개의 출력 행렬을 만드는 지, 
- 입력 채널 수란 들어온 입력 데이터의 채널 수를 뜻한다.
- 높이와 너비는 행과 열을 뜻한다.

- 배치형태의 계산을 위해 입력 데이터도 4차원으로 늘렸다.( 배치데이터수, 채널수, 행데이터, 렬데이터)


- 풀링 계층은 기본적으로 윈도우의 크기와 스트라이드 크기를 동일하게 한다.
- 풀링은 최대 풀링, 평균 풀링등이 있다.
- n*n에의 최대나 평균을 값으로 스트라이드만큼 움직인 후 작은 정방행렬을 만드는 것이다.
- 풀링은 채널 수에는 영향이 없다. 기본적으로 채널을 동일하게 만들어 준다.


- 풀링을 하는 이유 합성곱의 채널이 많고 이미지 데이터가 많다면 특징이 너무나 많아진다.
- 이런 경우 풀링을 이용해서 overfitting을 방지한다.

- 대표적인 CNN은 LeNet 과 AlexNet 이 있다.
- AlexNet은 활성화 함수로 ReLU를 사용하고, 배치 정규화와 드롭아웃을 사용한다.

- 신경망의 층이 깊어질수록 적은 매개변수로 같은 값을 표현할 수 있다.
- 매개변수가 적다는 것은 넓은 수용 영역을 소화할 수 있고 각 층마다 단계적으로 학습할 수 있다는 것에 있다.
- 첫 층에서는 특징을 추출하고 두 번째 층에서는 다른 과정을 거치며 복잡한 과정을 단순하게 풀어낼 수 있다.
