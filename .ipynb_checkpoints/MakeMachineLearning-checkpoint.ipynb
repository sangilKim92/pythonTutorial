```python
#퍼셉트론이란 주어진 입력값들이 가중치에 곱해졌을 때 임계치를 넘은 경우 출력하고 안넘은 경우 출력 안하는 것이다.
#이런 임계치와 가중치는 기존에는 인간이 정했지만 머신러닝으로 기계가 설정하게 하는것을 학습이라고 한다.
#ex) and 논리회로퍼셉트론을 파이썬으로 구현하기

def AND(x1,x2):
    w1, w2, theta = 0.5,0.5,0.7 #가중치, 가중치, 임계값
    tmp = x1*w1+x2*w2
    if tmp>=theta:
        return 1   
    elif tmp<theta:
        return 0
```

```python
print(AND(1,1))
print(AND(1,0))
print(AND(0,0))
print(AND(0,1))
```

```python
def AND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    tmp = np.sum(x*w)+b
    if tmp>=0:
        return 1
    elif tmp<0:
        return 0
        
def OR(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.2
    tmp=np.sum(x*w)+b
    if tmp>=0:
        return 1
    elif tmp<0:
        return 0
    
def NAND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.7
    tmp = np.sum(x*w)+b
    if tmp>=0:
        return 1
    elif tmp<0:
        return 0
```

```python
#위와 같이 퍼셉트론으론 and, or, nand에 대해선 선형방정식으로 쉽게 구현할 수 있음
#하지만 nor같은 배타적 논리합 구현을 못하였는데 다중 퍼셉트론 도입으로 가능해졌음
# nor은 비선형, and,or,nand는 선형
# nor은 and 와 or ,nand의 조합으로 만들 수 있다.

def XOR(x1,x2):
    s1=OR(x1,x2)
    s2=NAND(x1,x2)
    s3=AND(s1,s2)
    return s3
    
print(XOR(1,1))
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
```

```python
#위의 가중치를 일일이 설정하는것이 굉장히 귀찮고 힘든일임
#이런 가중치 설정을 컴퓨터가 학습하게 해주는 것이 신경망
#활성화 함수란 입력신호의 총합을 출력 신호로 변환하는 함수를 말한다.
#위의 식을 예로 보자면 AND함수내에서 0보다 큰경우 1을 출력, 0보다 작은 경우 0을 출력을 뜻한다.
#가중치와 편향은 따로 분류한다.
#단순 퍼셉트론에선 활성화함수가 출력1과 출력0과 같은 계단식이다. 하지만 다층 퍼셉트론은 시그모이드함수를 도입해서 활성화함수를 나타낸다.
#시그모이드함수는 입력값에 따라 값의 차이가 크지 않고 그것을 다음층으로 보낸다.

#계단함수를 간단하게 구현해보자
def step_function(x):
    if x>=0:
        return 1
    elif x<0:
        return 0

#앞으로는 numpy배열을 입력값으로 받고 그것에 대한 계단함수를 주로 사용하기에
def step_function(x):
    y=x>0
    return y.astype(np.int)#numpy타입 변환 함수 astype numpy는 위의 y=x>0 같이 하면 배열 하나하나 큰 경우에는 True 아니면 False
                            #를 가져온다. 그것을 타입 변환시켜 True는 1로 False 는 0으로 바꾸는 것이다.
    
def step_funtion(x):
    return np.array(x>0,dtype=np.int)

x=np.arange(-5.0,5.0,0.1)
y= step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
#시그모이드함수 구현 과 plt으로 보기 인수로 넘파이배열을 보내도 문제없이 된다. 넘파이배열은 스칼라값에 대해 사칙연산을 각각하기 때문에
#신경망에서는 활성화함수로 비선형 함수를 사용한다. 이유로는 선형함수는 신경망의 은닉층이 필요가 없어지기 때문이다.
#선형함수는 아무리 많이 쌓여도 하나의 선형함수로 표현이 가능하다.
#계단식함수와 시그모이드함수의 공통점은 우선 비선형함수라는 것에 있다. 그리고 0에서부터 1까지 값을 가진다는 점이 있다.
#다른 점이라면 시그모이드함수는 값에 따라 가까워지기도 하고 멀어지기도 하며 중요도를 나타낼수 있다는 점이다.
def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.arange(-5.0,5.0,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)#y축 범위 설정
plt.show()
```

```python
#신경망에서 시그모이드는 오래사용되었으나 최근에는 ReLU함수가 주로 사용된다.
#ReLU(Rectified Linear Unit) 함수로 0이하면 0을 출력하고 0이상이면 x값을 그대로 출력하는 간단한 함수이다.

def ReLU(x):
    return np.maximum(0,x)
#0부터 x까지의 최대값을 return하니 0이 음수라면 0을 출력하고 0이상이면 x를 출력한다.
```
