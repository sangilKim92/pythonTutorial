```python
#퍼셉트론이란 주어진 입력값들이 가중치에 곱해졌을 때 임계치를 넘은 경우 출력하고 안넘은 경우 출력 안하는 것이다.
#이런 임계치와 가중치는 기존에는 인간이 정했지만 머신러닝으로 기계가 설정하게 하는것을 학습이라고 한다.
#ex) and 논리회로퍼셉트론을 파이썬으로 구현하기

def AND(x1,x2):
    w1, w2, theta = 0.5,0.5,0.7 #가중치, 가중치, 임계값
    tmp = x1*w1+x2*w2
    if tmp>=theta:
        return 1   
    elif tmp<theta:
        return 0
```

```python
print(AND(1,1))
print(AND(1,0))
print(AND(0,0))
print(AND(0,1))
```

```python
def AND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    tmp = np.sum(x*w)+b
    if tmp>=0:
        return 1
    elif tmp<0:
        return 0
        
def OR(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.2
    tmp=np.sum(x*w)+b
    if tmp>=0:
        return 1
    elif tmp<0:
        return 0
    
def NAND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.7
    tmp = np.sum(x*w)+b
    if tmp>=0:
        return 1
    elif tmp<0:
        return 0
```

```python
#위와 같이 퍼셉트론으론 and, or, nand에 대해선 선형방정식으로 쉽게 구현할 수 있음
#하지만 nor같은 배타적 논리합 구현을 못하였는데 다중 퍼셉트론 도입으로 가능해졌음
# nor은 비선형, and,or,nand는 선형
# nor은 and 와 or ,nand의 조합으로 만들 수 있다.

def XOR(x1,x2):
    s1=OR(x1,x2)
    s2=NAND(x1,x2)
    s3=AND(s1,s2)
    return s3
    
print(XOR(1,1))
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
```
