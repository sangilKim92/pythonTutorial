```python
#새롭게 나온 시각화 라이브러리 위의 matplotlib에 비해 간단하게 사용된다.
import seaborn as sns

#class로 구별하도록 설정
sns.pairplot(df, hue="class")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(7, 7))
sns.set(style='white', palette='muted')

#seaborn 같은 경우 x축과 y축 두개를 다 설정해야 한다.
sns.violinplot(x=df['class'], y=df['sepal length'], ax=ax[0,0])
sns.violinplot(x=df['class'], y=df['sepal width'], ax=ax[0,1])
sns.violinplot(x=df['class'], y=df['petal length'], ax=ax[1,0])
sns.violinplot(x=df['class'], y=df['petal width'], ax=ax[1,1])
fig.suptitle('Violin Plots', fontsize=16, y=1.03)

#기존의 matplotlib에서 [0,0],[0,1],..,로 설정하던걸 for문으로 간단하게 구현
for i in ax.flat:
    plt.setp(i.get_xticklabels(), rotation=-90)
fig.tight_layout()
```

```python
#pandas에는 map, apply등 값을 찾아 시리즈(row)를 반환하는 함수를 제공한다.
df['class'] = df['class'].map({'Iris-setosa':'SET', 'Iris-virginica': 'VIR', 'Iris-versicolor': 'VER'})
#딕셔너리 기능은 map함수의 유일한 기능이고 apply대신 map을 선택한 이유가 단일 칼럼 변환 때문이다.


df
```

```python
#petal width의 평균은 1.3이었다. petal width 칼럼에 있는 값을 기준으로 데이터프레임에 wide petal이라는 새로운 칼럼 생성
#eptal width가 평균보다 크거나 같으면 1이라고 하고 평균 미만이면 0이라고 한다.
df['wide petal'] = df['petal width'].apply(lambda v: 1 if v >= 1.3 else 0)
#apply 함수는 칼럼과 시리즈 둘다 다룰 수 있다.


df
```

```python
#위의 함수는 단일 시리즈에 적용하였다. 이번에는 전체 데이터프레임에 적용해서 새롭게 칼럼에 추가하는것을 해본다.
df['petal area'] = df.apply(lambda r: r['petal length'] * r['petal width'], axis=1)
df
```
